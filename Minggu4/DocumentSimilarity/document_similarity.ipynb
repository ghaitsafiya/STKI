## **Document Similarity (Cosine Similarity)**
Menentukan seberapa mirip dua dokumen teks berdasarkan hasil representasi TF-IDF.

## Import Library
cosine_similarity digunakan untuk menghitung kesamaan antar-vektor teks (0 = tidak mirip, 1 = identik).

```python
# Load hasil TF-IDF
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load hasil TF-IDF dari file CSV
tfidf_df = pd.read_csv("tfidf_matrix.csv")

# Hitung Cosine Similarity dari data CSV
similarity_matrix = cosine_similarity(tfidf_df.values)
```

## Hitung Cosine Similariy
sim_df[i][j] = tingkat kemiripan antara dokumen i dan j.

Nilai mendekati 1 artinya dokumen-dokumen itu memiliki kata dan konteks mirip.

```python
# Tampilkan dalam bentuk tabel
sim_df = pd.DataFrame(
    similarity_matrix,
    index=[f"Doc{i+1}" for i in range(len(tfidf_df))],
    columns=[f"Doc{i+1}" for i in range(len(tfidf_df))])
sim_df
```
